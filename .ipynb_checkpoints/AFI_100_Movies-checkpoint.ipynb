```python
#Important libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd

#Fetching a given URL
url="https://en.wikipedia.org/wiki/AFI%27s_100_Years...100_Movies"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
soup
```

```python
tdata=soup.find(class_="wikitable sortable")
tdata
```

```python
heads=[g.get_text().strip() for g in tdata.find_all('th')]
heads
```

```python
res=[[g.get_text().strip() for g in t.find_all('td')] for t in tdata.find_all('tr')[1:]]
res
```

```python
d={k:[] for k in heads}
for p in res:
    d['Film'].append(p[0])
    d['Release year'].append(p[1])
    d['Director'].append(p[2])
    d['Production companies'].append(p[3])
    d['Rank'].append(p[4])
df=pd.DataFrame(d)
df
```

```python
df.to_csv("AFI_100_Movies.csv",index=False)
```
